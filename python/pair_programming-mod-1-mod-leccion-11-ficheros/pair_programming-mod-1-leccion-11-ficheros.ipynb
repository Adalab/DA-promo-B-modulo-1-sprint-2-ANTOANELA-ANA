```python
# Ejercicio 1: crear una funci贸n que me diga el directorio, cree dos carpetas una dentro de la otra y me mueva de una carpeta a otra. 

def ficheros(carpeta1, carpeta2, carpeta3):
    
    mi_directorio = os.getcwd()
    print(mi_directorio)

    contenido = os.listdir()

    if carpeta1 not in contenido:
        os.mkdir(carpeta1)
        print(f"Se ha creado la carpeta{carpeta1}")

    else:
        print(f"La carpeta {carpeta1} ya existe")

    os.chdir(carpeta1)
    print(os.getcwd())

    contenido_carpeta1 = os.listdir()

    if carpeta2 not in contenido_carpeta1:

        os.mkdir(carpeta2)

    else:
        print(f'La carpeta {carpeta2} ya existe')


    os.chdir(carpeta2)
    print(os.getcwd())

    os.chdir(mi_directorio)
    print(os.getcwd())

    os.rename(carpeta1, carpeta3)

    
```

```python
# Ejercicio 4: crear una funci贸n que tome como 煤nico par谩metro un archivo xml, me diga sus tags y attributes y finalmente me diga el t铆tulo, a帽o y descripci贸n 
# de cada pel铆cula en el archivo. 

import xml.etree.ElementTree as ET

def peliculas(archivo_peliculas):

    tree = ET.parse(archivo_peliculas)

    root = tree.getroot()
    print(root.tag)
    print(root.attrib)

    for child in root:
       
        print(child.tag)
        print(child.attrib)

        for subchild in child:

            print(subchild.tag)
            print(subchild.attrib)

            for subsubchild in subchild:

                print(subsubchild.tag)
                print(subsubchild.attrib['title'])
                print(subsubchild.find('year').text)
                print(subsubchild.find('description').text)


```

```python
peliculas(directorio_peliculas)
```
